```python
client = ors.Client(key='5b3ce3597851110001cf6248f4b8ded8bdf44b46bcb32ea1d9b1a2c4')
```

```python
#Longitude/Latitude of start point
x1=-0.337232
y1=51.421941
#Longitude/Latitude of end point
x2=6.799069
y2=45.890519
#Note that google maps uses Lat/Long (y,x) but ORS accepts only Long/Lat (x,y)
   
coords = [[x1,y1], [x2,y2]]

#Make API call to ORS
route = client.directions(coordinates=coords, 
                          profile='cycling-road',
                          format='geojson')

#Set map style and centre over the route
m = folium.Map(location=list(reversed([(x1+x2)/2, (y1+y2)/2])), tiles="cartodbpositron")

#Plot the route
folium.PolyLine(locations=[list(reversed(coord)) for coord in route['features'][0]['geometry']['coordinates']], color="blue").add_to(m)

#Set the bounds of the route
route_coords = pd.DataFrame(route['features'][0]['geometry']['coordinates'], columns=['Long', 'Lat'])
sw = route_coords[['Lat', 'Long']].min().values.tolist()
ne = route_coords[['Lat', 'Long']].max().values.tolist()
#sw = [x - 0.1 for x in sw]
#ne = [x + 0.1 for x in ne]

m.fit_bounds([sw, ne]) 

#Display the route
m
```
